# Line Charts

```python
df_store = pd.read_csv('Sample - Superstore.csv')
```

```python
df_store
```

#### `plt.plot`

### Date Handling

```python
df_store.info()
```

#### `pd.to_datetime()`

```python
df_store['Order Date'] = pd.to_datetime(df_store['Order Date'])
```

```python
df_store.info()
```

```python
df_store.sample(2).T
```

### Extract Year,Year-Month, Week-Year etc.

#### `dt.strftime()`

```python
df_store['Order_Year'] = df_store['Order Date'].dt.strftime('%Y')
```

```python
df_store['Order_Year-Month'] = df_store['Order Date'].dt.strftime('%Y-%m')
```

```python
df_store.sample(2).T
```

- Aggregating and sorting data for line chart

#### Filter the data only for last two years

```python
filt_year = df_store['Order_Year']>='2017'
```

```python
df_store_agg = df_store[filt_year].groupby('Order_Year-Month').agg(
Total_Sales = ('Sales', 'sum'),
Total_Profit = ('Profit', 'sum')).reset_index()
```

```python
df_store_agg
```

#### Themes in Matplotlib

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

#### We need a canvas

```python
plt.subplots(figsize=(16,6))
plt.show()
```

```python
plt.subplots(figsize=(16,6))

### Plot the line chart
plt.plot(df_store_agg['Order_Year-Month'], df_store_agg['Total_Sales'], linewidth=2, marker='o', label='Total Sales Trend', color='purple')
plt.xticks(rotation=45)
plt.xlabel('Order Year Month')
plt.ylabel('Total Sales in $')
plt.title('Sales Trend Month over Month')
plt.legend()
plt.show()
```

### Add a trend chart for Profit

```python
plt.subplots(figsize=(16,6))

### Plot the line chart  - Sales
plt.plot(df_store_agg['Order_Year-Month'], df_store_agg['Total_Sales'], linewidth=2, marker='o', label='Total Sales Trend', color='purple')

### Plot the line chart - Profit
plt.plot(df_store_agg['Order_Year-Month'], df_store_agg['Total_Profit'], linewidth=2, marker='o', label='Total Profit Trend', color='green')

plt.xticks(rotation=45)
plt.xlabel('Order Year Month')
plt.ylabel('Total Sales in $')
plt.title('Sales Trend Month over Month')
plt.legend()
plt.show()
```

# Bar Chart

- A barplot or bar chart is one of the most popular charts
- shows a relationship between a numeric and a categorical varibale
- each entity of the categorical variable represents as a bar
- size of the bar represents numeric value

### Show top 5 sub-categories by `Total Sales`

```python
df_store_subcat_agg = df_store.groupby('Sub-Category').agg(
Total_Sales = ('Sales', 'sum')).reset_index()
```

```python
df_store_subcat_agg
```

```python
df_store_subcat_agg_sorted = df_store_subcat_agg.sort_values(by='Total_Sales', ascending=False)
```

```python
df_store_subcat_agg_sorted
```

```python
df_store_subcat_agg_sorted_top5 = df_store_subcat_agg_sorted.head(5)
```

```python
df_store_subcat_agg_sorted_top5
```

#### `plt.bar()`

```python
plt.subplots(figsize=(8,4))

### Plot the bar chart
plt.bar(df_store_subcat_agg_sorted_top5['Sub-Category'], df_store_subcat_agg_sorted_top5['Total_Sales'], width=0.3, color='Orange')

plt.xticks(rotation=45)
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales in $')
plt.title('Top 5 Selling Sub-Categories')
# plt.legend()
plt.show()
```

### `dash plotly` , `Django`, `flask`

```python
plt.subplots(figsize=(8,4))

### Plot the bar chart
plt.barh(df_store_subcat_agg_sorted_top5['Sub-Category'], df_store_subcat_agg_sorted_top5['Total_Sales'], color='Orange')

plt.xticks(rotation=45)
plt.xlabel('Total Sales in $')
plt.ylabel('Sub-Category')
plt.title('Top 5 Selling Sub-Categories')
# plt.legend()
plt.show()
```

# Pie Chart

- Students marks displayed as pie chart
- Its a circular statistical graphic
- Used to show distribution of total

```python
df_store_cat = df_store.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
```

```python
df_store_cat
```

```python
plt.style.available
```

```python
plt.style.use('tableau-colorblind10')
```

#### `plt.pie`

```python
plt.subplots(figsize=(5,5))

### Plot the pie chart
plt.pie(df_store_cat['Sales'], labels=df_store_cat['Category'], autopct='%1.2f%%', startangle=90)


plt.show()
```

```python
plt.subplots(figsize=(5,5))

### Plot the pie chart
explode1 = [0,0,0.1]
plt.pie(df_store_cat['Sales'], labels=df_store_cat['Category'], autopct='%1.2f%%', startangle=90, explode=explode1)


plt.show()
```

# Scatter Plot

- To show the relationship between a pair of numerical variables
- Used in Linear Regression
- Plots the data on horizontal and vertical axes in order to show how much one variable is affected by the other

```python
plt.subplots(figsize=(15,6))

### Plot the scatter chart
plt.scatter(df_store['Sales'], df_store['Profit'], color='green', s=85)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Relationship between Sales and Profit')

plt.show()
```
