## Why use `for` loop?

Lets see the use **`for`** loop in Python.

* **Definite Iteration:** When we know how many times we wanted to run a loop, then we use count-controlled loops such as **`for`** loops. It is also known as definite iteration. For example, Calculate the percentage of 50 students. here we know we need to iterate a loop 50 times (1 iteration for each student).
* **Reduces the codes complexity:** Loop repeats a specific block of code a fixed number of times. It reduces the repetition of lines of code, thus reducing the complexity of the code. Using **`for`** loops and while loops we can automate and repeat tasks in an efficient manner.
* **Loop through sequences:** used for iterating over lists, strings, tuples, dictionaries, etc., and perform various operations on it, based on the conditions specified by the user.

* **Indexing and Enumeration:** `for` loop can provide access to the index or position of each item in a sequence, which can be useful for tracking the order of items of for reference.

* **Searching and matching:** `for` loop can also be used to search specific elements or patterns in data


```python
nums = [1,2,3,4,5]

for i in nums:
    print(i*2)
```

## Conditional Loops: `for` loop with `if-else`

A **`for`** loop can have an optional [if-else] block. The **`if-else`** checks the condition and if the condition is **`True`** it executes the block of code present inside the **`if`** block and if the condition is **`False`**, it will execute the block of code present inside the **`else`** block.

### Use Case: Categorizing Student Grades

Suppose we have a list of 10 student grades and need to categorize into dfiferent grade as per the criteria below:
    
    - A: 90+
    - B: 80-90
    - C: 65-80
    - D: 55-65
    - F: < 55
    
   `grades = [75,40,95,88,65,52,59,82,92,65]`
   
   **Store the output in the form of dictionary: {75:'Grade-C'}**

```python
grades = [75,40,95,88,65,52,59,82,92,65]
```

```python
grades = [75, 40, 95, 88, 65, 52, 59, 82, 92, 65]

for grade in grades:
    if grade >= 90:
        print(f"Grade {grade}: A")
    elif grade >= 80:
        print(f"Grade {grade}: B")
    elif grade >= 65:
        print(f"Grade {grade}: C")
    elif grade >= 55:
        print(f"Grade {grade}: D")
    else:
        print(f"Grade {grade}: F")
```

### APC

```python
### Marks of students in Python Test
marks = [75,40,95,88,75,52,59,82,92,65]

### Dictionary defining the grade criteria
grade_criteria = {
    "A":(90,100),
    "B":(80,89),
    "C":(65,79),
    "D":(55,64),
    "F":(0,54)
}

### Create an empty dictionary to store the grade categories
grade_categories = {}

### Iterate through the list of marks
for score in marks:
    for category, (min_score, max_score) in grade_criteria.items():
        if min_score <= score <=max_score:
            grade_categories[score] = category
            break 
      
```
