#### If the state of the object can't change after the object has been constructed, it is immutable
`For example: strings, tuples are immutable`

#### If the state of the object can be modified, it is mutable
`For example: list, dictionary are mutable`

## Functions vs Methods

#### Functions: They are defined using the `def` keyword followed by a function name, list of parameters in (), and a colon to indicate start of the function block. 
- User defined function, typically defined at the module level which can be used globally within the module

#### Methods are functions that are associated objects or classes. They are defined wthn the scope class using `def` keyword and a special parameter called `self`

**In summary, functions are standalone piece of code that can be called independently while methods are functions that are associated with objects or classes and are called on instances of those classes**
