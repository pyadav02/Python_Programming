## Why use `for` loop?

Let’s see the use **`for`** loop in Python.

* **Definite Iteration:** When we know how many times we wanted to run a loop, then we use count-controlled loops such as **`for`** loops. It is also known as definite iteration. For example, Calculate the percentage of 50 students. here we know we need to iterate a loop 50 times (1 iteration for each student).
* **Reduces the code’s complexity:** Loop repeats a specific block of code a fixed number of times. It reduces the repetition of lines of code, thus reducing the complexity of the code. Using **`for`** loops and while loops we can automate and repeat tasks in an efficient manner.
* **Loop through sequences:** used for iterating over lists, strings, tuples, dictionaries, etc., and perform various operations on it, based on the conditions specified by the user.

* **Indexing and Enumeration:** `for` loop can provide access to the index or position of each item in a sequence, which can be useful for tracking the order of items of for reference.

* **Searching and matching:** `for` loop can also be used to search specific elements or patterns in data


```python
nums = [1,2,3,4,5]

for i in nums:
    print(i*2)
```

The **`if-else`** checks the condition and if the condition is **`True`** it executes the block of code present inside the **`if`** block and if the condition is **`False`**, it will execute the block of code present inside the **`else`** block.

### Use Case: Categorizing Student Grades

Suppose we have a list of 10 student grades and need to categorize into dfiferent grade as per the criteria below:
    
    - A: 90+
    - B: 80-90
    - C: 65-80
    - D: 55-65
    - F: < 55
    
   `grades = [75,40,95,88,65,52,59,82,92,65]`
   
   **Store the output in the form of dictionary: {75:'Grade-C'}**

```python
grades = [75,40,95,88,65,52,59,82,92,65]
```

```python
grades = [75, 40, 95, 88, 65, 52, 59, 82, 92, 65]

for grade in grades:
    if grade >= 90:
        print(f"Grade {grade}: A")
    elif grade >= 80:
        print(f"Grade {grade}: B")
    elif grade >= 65:
        print(f"Grade {grade}: C")
    elif grade >= 55:
        print(f"Grade {grade}: D")
    else:
        print(f"Grade {grade}: F")
```

### APC

```python
### Marks of students in Python Test
marks = [75,40,95,88,75,52,59,82,92,65]

### Dictionary defining the grade criteria
grade_criteria = {
    "A":(90,100),
    "B":(80,89),
    "C":(65,79),
    "D":(55,64),
    "F":(0,54)
}

### Create an empty dictionary to store the grade categories
grade_categories = {}

### Iterate through the list of marks
for score in marks:
    for category, (min_score, max_score) in grade_criteria.items():
        if min_score <= score <=max_score:
            grade_categories[score] = category
            break 
      
```

## Continue 

- it is used to stop/skip the block of the code in the `for` loop for the **current iteration only** and continue with the next iteration

```python
nums = [10,20,30,40,60,150,45,65,33,105,90,99, 100, 7]

for i in nums:
    if i >=100: #condition to just skip the current iteration only if the condition is satisfied
        continue
    print('Current number is:',i)
    
    
print('Done')
```

### What is a function in Python?

In Python, a **function is a block of organized, reusable (DRY- Don’t Repeat Yourself) code with a name** that is used to perform a single, specific task. It can take arguments and returns the value.

Functions help break our program into smaller and modular chunks. As our program grows larger and larger, functions make it more organized and manageable.

Furthermore, it improves efficiency and reduces errors because of the reusability of a code.

### Types of Functions

1. Built-in function

Q. Using hours and rate (rate per hour), calculate $ amount to be paid to the employees as grosspay. Take into account that    the factory gives the employee `1.5 times the hourly rate` for hours worked above 40 hours (Weekly hours cut-off). Write a function called `computepay` which takes `two` parameters namely: `hours` and `rate`

**Hourly rate: $30/hour**


```python
def computepay(hours, rate):
    if hours <= 40:
        gross_pay = hours * rate
    else:
        regular_pay = 40 * rate
        extra_pay = (hours - 40) * (1.5 * rate)
        gross_pay = regular_pay + extra_pay
    return gross_pay

gross_pay = computepay(50, 30)
print(gross_pay)
```

```python
def pay(hours, rate):
    if hours > 40:
        regular_pay = 40 * rate
        overtime_pay = (hours - 40) * (rate * 1.5)
        total_pay = regular_pay + overtime_pay
    else:
        total_pay = hours * rate

    return total_pay

total_pay = pay(50, 30)
print("Total pay: ", total_pay)
```

```python
def given_operation(a, b, operation='add'):
    if operation == 'add':
        return a + b
    elif operation == 'sub':
        return a - b
    elif operation == 'mul':
        return a * b
    elif operation == 'div':
        if b == 0:
            return "Cannot divide by zero"
        return a / b
    elif operation == 'fdv':
        if b == 0:
            return "Cannot perform floor division by zero"
        return a // b
    elif operation == 'mod':
        if b == 0:
            return "Cannot perform modulus by zero"
        return a % b
    elif operation == 'exp':
        return a ** b
    else:
        return "Invalid operation"
```

```python
given_operation(100,200, 'apc')
```

```python
def calculator(value1,value2,operation='+'):
    
    if operation =='+':
        result=value1+value2
    elif operation =='-':
        result=value1-value2
    elif operation =='*':
        result=value1*value2
    elif operation =='/':
        result=value1/value2
    elif operation =='%':
        result=value1%value2
    return result
```

```python
def do_math(x, y, operation='add'):
    if operation == 'add':
        return x + y
    elif operation == 'sub':
        return x - y
    elif operation == 'mul':
        return x * y
    elif operation == 'div':
        if y != 0:
            return x / y
        else:
            return "Cannot divide by zero"
    elif operation == 'fdiv':
        return x // y
    elif operation == 'mod':
        return x % y
    elif operation == 'exp':
        return x ** y
    else:
        return "Invalid operation"

```

### `filter()`

- is used to generate an output list of values that return true when the function is called

filter(function, iterables)

```python
numbers = list(range(1,101))
```

```python
def is_even(x):
    return x%2==0 #remainder is 0 implies even function
```

```python
### Use the filter function to get a list of even numbers
even_nums = filter(is_even, numbers)
```

```python
even_nums
```

```python
list(even_nums)
```

```python
### Use the filter function to get a list of even numbers
even_nums = map(is_even, numbers) 
```

```python
len(list(even_nums))
```

### `map` computes & `filter` extracts as per condition

In summary, filter is used to select elements from an iterable **based on a condition**, while map is used to apply a function to each element of an iterable. 
The choice between filter and map depends on the **specific task** you need to perform on the elements of the iterable.
