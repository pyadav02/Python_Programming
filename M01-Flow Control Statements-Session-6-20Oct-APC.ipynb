### Nested if statement with `user input`

**ONLY USE `IF` STATEMENT**

```python
score = int(input('Enter a valid score:'))

if score >=300:
    print("Triple Ton")
if 200<=score <300:
    print("Double Ton")
if 100<= score <200:
    print("Single Ton")
if score<100:
    print("No Ton")
```

### Using Nested if

```python
score = int(input('Enter a valid score:'))

if score >=300:
    print("Triple Ton")
if score < 300:
    if score >=200:
        print("Double Ton")
if score <200:
    if score>=100:
        print("Single Ton")
if score <100:
    print("No Ton")
        
```
